```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset.head(2)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

```python
dataset["Order Priority"] = dataset["Order Priority"].replace("Crititcal ", "Crititcal")
```

```python
dataset['Order Priority'].value_counts()
```

```python
dataset["Order Priority"] = dataset["Order Priority"].replace("Crititcal", "Critical")
```

```python
dataset['Order Priority'].unique()
```

# ORDER PRIORITY

```python
dataset['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority", data=dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

# SHIPPING MODE

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values

plt.pie(y,labels=x,startangle = 60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x="Ship Mode",data=dataset,hue="Product Category")
plt.show()
```

# CUSTOMER SEGMENT

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Segment",data=dataset,)
plt.show()
```

# Product Category	

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Product Category",data=dataset,hue="Product Sub-Category")
plt.legend(bbox_to_anchor = (1.00, 1.05), )
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Product Category",data=dataset[dataset["Product Category"]=="Office Supplies"],hue="Product Sub-Category")
plt.legend(bbox_to_anchor = (1.00, 1.05), )
plt.show()
```
